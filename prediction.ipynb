```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
###Load all the trained model,Scalar pickle file, and one hot encoder pickle file
model=load_model('model.h5')

##Load encoders and scalar
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender= pickle.load(file)
with open('one_hot_encoder_geo.pkl','rb') as file:
    one_hot_encoder_geo= pickle.load(file)
with open('scalar.pkl','rb') as file:
    scalar= pickle.load(file)
```

```python
input_data={
    'CreditScore':600,
    "Geography":'France',
    'Gender':'Male',
    'Age':45,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':55000
}
```

```python
geo_encoded = one_hot_encoder_geo.transform([[input_data['Geography']]])
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
df=pd.DataFrame([input_data])
df
```

```python
df=df.drop(columns='Geography')
```

```python
df=pd.concat([df,geo_encoded_df],axis=1)
```

```python
df
```

```python
##Encode Categorical variable
df['Gender']=label_encoder_gender.transform(df['Gender'])
```

```python
df
```

```python
## Scaling the data
scaled_df=scalar.transform(df)
scaled_df
```

```python
prediction=model.predict(scaled_df)
prediction
```

```python
prediction_prob=prediction[0][0]
```
